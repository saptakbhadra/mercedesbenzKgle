```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_csv('/content/drive/My Drive/Datasets/MercedesBenzDatasets/train.csv')
test_df=pd.read_csv('/content/drive/My Drive/Datasets/MercedesBenzDatasets/test.csv')
```

```python
pd.pandas.set_option('display.max_columns', None)
pd.pandas.set_option('display.max_rows', None)
```

```python
print('train_df' , train_df.shape , 'test_df',test_df.shape)
```

```python
y_train=train_df['y'].values
id_test=test_df['ID'].values
```

```python
train_copy=train_df.copy()
test_copy=test_df.copy()
```

```python
train_copy.drop(['y','ID'],axis=1,inplace=True)
test_copy.drop(['ID'],axis=1,inplace=True)
train_copy.head()
```

```python
final_df=pd.concat([train_copy,test_copy],axis=0,sort=False)
```

```python
final_df.shape
```

```python
from sklearn import preprocessing
for f in ['X0','X1','X2','X3','X4','X5','X6','X8']:
    lbl=preprocessing.LabelEncoder()
    lbl.fit(list(final_df[f].values))
    final_df[f]=lbl.transform(list(final_df[f].values))
```

```python
final_df.head()
```

```python
corr=final_df.corr()
```

```python
range(corr.shape[0])
```

```python
columns = np.full((corr.shape[0]), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
selected_columns = final_df.columns[columns]
final_df = final_df[selected_columns]
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
X_train=final_df.iloc[:4209]
X_test=final_df.iloc[4209:]
```

```python
import xgboost
regressor=xgboost.XGBRegressor()

booster =['gbtree' , 'gblinear']
base_score=[0.25,0.5,0.75,1]

## Hyper parameter Optimization

n_estimators = [100,500,900,1100,1500]

max_depth=[2,3,5,10,15]
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

# Define the grid of hyperparameters to search
hyperparameter_grid={
    'n_estimators':n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }

# Set up the random search with 4-fold cross validation
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'r2',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)


random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
regressor=xgboost.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0,
             importance_type='gain', learning_rate=0.1, max_delta_step=0,
             max_depth=2, min_child_weight=1, missing=None, n_estimators=100,
             n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=1, verbosity=1)

regressor.fit(X_train,y_train)
```

```python
import pickle
filename='xgbhyper_model.pkl'
pickle.dump(regressor,open(filename,'wb'))
```

```python
prediction_xgb=regressor.predict(X_test)
```

```python
sub=pd.DataFrame()
sub['ID']=id_test
sub['y']=prediction_xgb
sub.to_csv('/content/drive/My Drive/Datasets/MercedesBenzDatasets/mylassoxgb.csv',index=False)
```
